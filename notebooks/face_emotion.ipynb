```python
def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def angle_between(landmarks):
    l1, l2, l3 = landmarks
    v1 = l1 - l2
    v2 = l3 - l2
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return np.degrees(np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0)))

def extract_features(landmarks, triplet_ids):
  triplet = landmarks[triplet_ids]
  return {
      "landmarks" : triplet,
      "angle" : angle_between(triplet)
  }

def extract_angles_only(landmarks, triplet_ids):
  triplet = landmarks[triplet_ids]
  return angle_between(triplet)
```

```python
def get_face_mesh(image_data):
  image_file = image_data
  # label = image_data[1]
  with mp_face_mesh.FaceMesh(
    static_image_mode=True,
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5) as face_mesh:
    
    image = cv2.imread(image_file)
    # Convert the BGR image to RGB before processing.
    results = face_mesh.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # Print and draw face mesh landmarks on the image.
    if not results.multi_face_landmarks:
      return
    normalized_landmarks = results.multi_face_landmarks[0].landmark
    face_landmarks = np.array([(normalized_landmarks[idx].x, normalized_landmarks[idx].y) for idx in LANDMARK_IDS])
    # face_landmarks = np.array([(landmark.x, landmark.y) for landmark in normalized_landmarks])
    angles = np.array([extract_angles_only(face_landmarks, np.array(triplet)) for triplet in ANGLE_IDS])
    # angles = np.array([triplet['angle'] for triplet in triplets])
    return angles
    # return {
    #     'landmarks' : face_landmarks,
    #     # 'triplets' : triplets,
    #     'angles' : angles,
    #     'label' : emotion_to_label[label]
    # }
```

```python
def softmax(x):
    x = x.reshape(-1)
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

def postprocess(scores):
    prob = softmax(scores)
    prob = np.squeeze(prob)
    classes = np.argsort(prob)[::-1]
    return classes

def detect_face(image_file: str):
  with mp_face_detection.FaceDetection(
    model_selection=1, min_detection_confidence=0.5) as face_detection:
  
    image = cv2.imread(image_file)
      # Convert the BGR image to RGB and process it with MediaPipe Face Detection.
    results = face_detection.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    if not results.detections:
      return "not result"

  return results

def get_face_from_detection(img, results):
  x_min = abs(results.detections[0].location_data.relative_bounding_box.xmin)
  y_min = abs(results.detections[0].location_data.relative_bounding_box.ymin)
  width = abs(results.detections[0].location_data.relative_bounding_box.width)
  height = abs(results.detections[0].location_data.relative_bounding_box.height)

  h, w, c = img.shape 
  face = img[round(h*y_min) : round(h*(y_min + height)), 
             round(w*x_min) : round(w*(x_min + width))]
  face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
  face = cv2.resize(face, dsize=(64,64), interpolation=cv2.INTER_CUBIC)
  face_data = np.resize(face, (1, 1, 64, 64))
  return face_data

    
def run(input_data):
    try:
        # load in our data, convert to readable format
        data = np.array(input_data).astype('float32')
        
        start = time.time()
        r = session.run([output_name], {input_name : data})
        end = time.time()
        
        # result = emotion_map(postprocess(r[0]))
        
        result_dict = {"result": postprocess(r[0]),
                      "time_in_sec": [end - start]}
    except Exception as e:
        result_dict = {"error": str(e)}
    return result_dict
    # return json.dumps(result_dict)

def classify_emotion(file_name):
    results = detect_face(file_name)
    if type(results) == str:
      return None
    image = cv2.imread(file_name)
    face = get_face_from_detection(image, results)
    emotion = run(face)
    return emotion
```
